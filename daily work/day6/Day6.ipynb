```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# 1) generating random weight data for 1000 people between 50 and 100 kg
np.random.seed(42)
weights = np.random.uniform(50, 100, 1000)
```

```python
# Plotting the histogram as thr PMF
plt.hist(weights, bins=np.arange(50,101,2), color='lightgray',edgecolor='black', alpha = 0.7, label='PMF', density=True)

plt.xlabel('Weight (KG)')
plt.ylabel('Probability mass')
plt.title('probability mass function for 1000 people')
plt.legend()
plt.show()
```

```python
# 2)PMF on the Age column of diabetes dataset

df=pd.read_csv("diabetes.csv")
age_data=df["Age"]
```

```python
plt.hist(age_data,bins="auto",color='hotpink',edgecolor='red',alpha=0.7,label="PMF")
#use "auto" for bins selection
plt.xlabel("Age")
plt.ylabel('probability mass')
plt.title("probablity mass function of age in the dataset")
plt.legend()
plt.show()
```

```python
# 3) Generate random weights data for 1000 people between 50 to 100 kg
np.random.seed(42)
weights = np.random.uniform(50,100,1000)
```

```python
# Filter weights withing a range 65 to 67kg for KDE visualiation
weights_kde_range = weights[(weights>=65) & (weights <= 67)]
```

```python
# Plotting the histogram for the entire weight range with density paramenter
plt.hist(weights, bins=np.arange(50,101,2), color = 'green', edgecolor = 'red', alpha=0.7, label='Overall Distribution', density=True )

# plotting the KDE (PDF) for the weights data in the specified  range with commom_norm
sns.kdeplot(weights_kde_range, bw_method=0.5, fill=True,color='orange', alpha=0.7,linewidth=2, label='KDE (65-67 KG)', common_norm=False)

# comma_norm=True (default): Total area under all the curves should be 1
# comma_norm=False: Area under our own curve should be 1

plt.xlabel('Weight (KG)')
plt.ylabel('Probability Density')
plt.title('Weight Distribution and KDE (65-67 KG) for 1000 people')
plt.legend()
plt.show()
```
