```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python

plt.hist(df.BloodPressure,bins='auto',color='yellow',edgecolor='red',alpha=1,label='overall distribution',density=True)
blood_range=df.BloodPressure[(df.BloodPressure>=40)&(df.BloodPressure<=50)]


sns.kdeplot(blood_range,bw_method=0.5,fill=True,color='hotpink',alpha=0.7,linewidth=2,label="kde (40-50)",common_norm=False)
plt.xlabel("blood pressure")
plt.ylabel("probablity density")
plt.title("blood pressure distrubution and ked (40-50)")
plt.legend()
plt.show()
```

```python


plt.hist(df.BloodPressure, bins='auto', color='pink', edgecolor='blue', alpha=0.8, label='Overall Distribution', density=True)
bg_range_kde=df.BloodPressure[(df.BloodPressure>=90)&(df.BloodPressure<=120)]

sns.kdeplot(bg_range_kde, bw_method=0.5, fill=True, color='red', alpha=0.9, linewidth=2, label='KDE (90-120)', common_norm=False)
plt.xlabel('Blood Pressure')
plt.ylabel('Probabiltity Density')
plt.title('Blood Pressure Distribution and KDE (90-120)')
plt .legend()
plt.show()
```
