```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data =pd.read_csv('tips.csv')
```

```python
#calculate descriptive statistics for the 'tip column
tip_stats=data['tip'].describe()
print('descriptive statistics for the "tip" column:')
print(tip_stats)
```

```python
#additional statistics
mean_tip=data.tip.mean()
median_tip=data.tip.median()
std_tip=data.tip.std()
var_tip=data.tip.var()
skew_tip=data.tip.skew()
kurtosis_tip=data.tip.kurtosis()
cv_tip=std_tip/mean_tip
```

```python
print("additional statistics for the tip column:")
print(f"mean:{mean_tip}\nMedian:{median_tip}\nstandart deviation:{std_tip}\nvariance:{var_tip}\nskewness:{skew_tip}\nkurtosis:{kurtosis_tip}\ncoefficient of variation:{cv_tip}%")
```

```python
#Ploting the distribution of the tip columns
plt.figure(figsize=(14,6))

#histogram
plt.subplot(1,2,1) #number of rows number of columns current subplot number
sns.histplot(data.tip,bins=20,kde=True,color="magenta")
plt.title("distribution of tips")
plt.xlabel("tip amount")
plt.ylabel("frequency")




#box plot 
plt.subplot(1,2,2)
sns.boxplot(y=data.tip)
plt.title('box plot of tips')
plt.xlabel ("top amount")
#show the plot 
plt.tight_layout()
plt.show()
```
