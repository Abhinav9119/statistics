```python
import pandas as pd
wdi=pd.read_csv('world_development_indicators.csv')
wdi.info()
```

```python
#colmn analusis
#some columns seem to be repeatind

print(wdi[['country_name','Country Name','Country Code','planet']].head())
print(wdi.nunique())
```

```python
#needed?
print(wdi['planet'].value_counts())
wdi['planet'].value_counts(normalize=True)
print(wdi[wdi['country_name']!=wdi['Country Name']])
print(wdi['country_name'].compare(wdi['Country Code']))
```

```python
#ok to drop these >...............
wdi=wdi.drop(columns=['Country Name','Country Code','planet'])
print(wdi.info())
```

```python
#row analysis
print(wdi.duplicated())
print(wdi[wdi.duplicated()])
print(wdi.drop_duplicates(ignore_index=True))
wdi=wdi.drop_duplicates(ignore_index=True)
wdi.info()
```

```python
#verify that duplidcates are gome
print(wdi.shape)
print(wdi[wdi.duplicated()])
```

```python
#missing data handling
import pandas as pd 
wdi = pd.read_pickle('wdi.pkl')
print(wdi.info()) #get the none - null count for each column

print(wdi['alcohol_consumption_per_capita'])
print(wdi.isna())
```

```python
#how many nulls in jeach column?
print(wdi.isna().sum())
```

```python
#unique counts for each value for the alcohol_consumption_per capita cloumn
#note :
#this excludes numms by default
print(wdi["alcohol_consumption_per_capita"].value_counts())
print(wdi['alcohol_consumption_per_capita'].value_counts(dropna=False))
```

```python
#find nulls by rows 
print(wdi.isna())
print(wdi.isna().sum(axis=1))
```

```python
#store this as a series
num_missing_by_row =wdi.isna().sum(axis=1)
print(num_missing_by_row>0)
print((num_missing_by_row>0).sum())
print(wdi.shape)
```

```python
print(wdi[num_missing_by_row>0])
```

```python
#drop rows that contain missing values
print(wdi.dropna())
print(wdi.dropna().info())
print(wdi.dropna(axis=1))
```
