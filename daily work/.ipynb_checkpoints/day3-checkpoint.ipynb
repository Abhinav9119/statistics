```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a= np.linspace(0,10,11)
b=a**4
print(b)
```

```python
x=np.arange(0,10)
y=2*x
print(x)
print(y)
```

```python
#the plt.subplots() object will act as a more automatic axis manager 
#this makes it much easier to show multiple plots side by side
fig,axes=plt.subplots()
axes.plot(x,y,"r")  #r=red color
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("title");  #; hides out[]
plt.show()


```

```python
""" then ypu can specify the number of rows and columns when creating the  subplots() object:"""

#empty canvas of 1 by 2 subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
print(axes)
print(axes.shape)
plt.show()
```

```python
#use .spbplots_adjusts to adjust spacing manually
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(12,8))
#parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)
plt.show()
```

```python
#use .spbplots_adjusts to adjust spacing manually
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=(12,8))
#parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)
fig.subplots_adjust(left=None,bottom=None,right=None,top=None,wspace=0.9,hspace=0.1)  #j
plt.show()
```

```python
# Example 2

x=np.arange(0,10)
y=2*x

#Legends
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')
ax.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, x**2, 'b.-' ,label='x**2')
ax.plot(x, x**3, 'g--', label='x**2')

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, x+1, color='blue', alpha=0.5)
ax.plot(x, x+2, color='#b2fc0a')
ax.plot(x, x+3, color='#ff6019')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, x-1, color='red', linewidth = 0.25)
ax.plot(x, x-2, color='red', lw = 0.50)
ax.plot(x, x-3, color='red', lw = 0.75)
ax.plot(x, x-4, color='red', lw = 10)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, x-1, color='green', lw = 3, ls = '--')
ax.plot(x, x-2, color='green', lw = 3, ls='-.')
ax.plot(x, x-3, color='green', lw = 3)
ax.plot(x, x-4, color='green', lw = 3)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(12,6))
#use marker

ax.plot(x,x-1,marker="+",markersize=20)
ax.plot(x,x-2,marker="o",ms=20)
ax.plot(x,x-3,marker="s",ms=20,lw=0)  #ms can be used for markersize
ax.plot(x,x-4,marker="1",ms=30,lw=15)  #make linewidth 
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

df=pd.read_csv("patient.csv")
ages =df.groupby("age").median().reset_index()
```

```python
#simple pandas
ages.plot.line("age","chol");
plt.show()
```

```python
#matplotlib
fig,ax=plt.subplots()
ax.plot(ages["age"],ages["chol"],ls=":",lw=1.7)
ax.set_xlabel("Age")
ax.set_ylabel("cholestrol");
plt.show()
```

```python
#seaborn
sns.lineplot(x="age",y="chol",data=ages,linestyle=":",linewidth=1.7)
plt.xlabel("age")
plt.ylabel("cholestrol")

plt.show()
```

```python
df=pd.read_csv("dm_office_sales.csv")
sns.scatterplot(x="salary",y="sales",data=df)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df)
plt.show()
```

```python
#hue:coloe points based off a categirical feature in the dataframe
sns.scatterplot(x="salary",y="sales",data=df,hue="work experience")
plt.show()
```

```python
#chosing a palette from matplotlibs cmap:
#https:/matplotlib.org/tutorials/colors/colormaps.html

plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df,hue="work experience",palette="spring")
plt.show()
```

```python

#use *s=* 
plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df,hue="work experience",palette="spring")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df,s=200)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df,s=200,linewidth=0,alpha=0.2)
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df,style='level of education')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x="salary",y="sales",data=df,style="level of education",hue="level of education",s=100)
plt.show()
```

```python
#count plot: a somple plot, it merely  shows the tptal countof rows per category 
#a histogram across a categorical , instead of qualitative ,variable
#categorical plot

plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x="level of education",data=df)
plt.show()
```

```python
#breakdown within another category with "hue"
plt.figure (figsize=(10,4),dpi=200)
sns.countplot(x='level of education',data=df,hue='training level')
plt.show()
```

```python
#using palette
plt.figure (figsize=(10,4),dpi=200)
sns.countplot(x="level of education",data=df,hue="training level",palette="Set1")
plt.show()
```

```python

plt.figure (figsize=(10,4),dpi=200)
#paired would be a good choice if there was a distinct jump from 0 and 2 and 3
sns.countplot(x="level of education",data=df,hue="training level",palette="Paired")
plt.show()
```
