```python
import pandas as pd
from scipy import stats
```

```python
data=pd.read_csv("Mall_customers.csv")
age_data=data['Age']
```

```python
#hypothesis population average age
pop_avg_age=40
```

```python
#prtform one sample t-test
t_statistic,p_value=stats.ttest_1samp(age_data,pop_avg_age)
```

```python
#degrees of fredom
degrees_of_fredom=len(age_data)-1
```

```python
#print the result
print("t statistics:",t_statistic)
print("p value:",p_value)
print("degrees of freedom:",degrees_of_fredom)
```

```python
#interpretation
alpha=0.05
if p_value<alpha:
    print("the null hypothesis (mean age=40) is rejected>>>")
else:
    print("the null hypothesis (mean age =40) cannot be rejected")

critical_value=stats.t.ppf(1-alpha,degrees_of_fredom)
print("critical value:",critical_value)
```

```python
#two tail t test (because we are just chedcking if sample mean =population mean)
#so we need to tale absolute value of t statistic
if (abs(t_statistic)<critical_value):
    print("the null hypothesis (mean age=40) cannot be rejected...")
else:
    print("the null hypothesis (mean age=40) is rejected")
```

```python
#two tail t test 
if (abs(t_statistic)<critical_value):
    print("the null hypothesis (mean finish time =130) cannot be rejected")
else:
    print("the null hypothesis (mean finish time =130) is rejected")
```

<h2>two independent samples t-test<h/2>

```python
from scipy import stats
```

```python
data=pd.read_csv('Mall_customers.csv')
male_spending=data[data['Genre']=="Male"]["Spending Score (1-100)"]
female_spending=data[data["Genre"]=='Female']["Spending Score (1-100)"]
```

```python
#perform two sample t test
t_statistic,p_value=stats.ttest_ind(male_spending,female_spending)
```

```python
#degree of freeedom 
n_male=len(male_spending)
n_female=len(female_spending)
degrees_of_freedom=n_male+n_female-2
```

```python
#siginficance level (alpha)
alpha =0.05
```

```python
#find th critical value
critical_value=stats.t.ppf(1-alpha/2,degrees_of_freedom)
```

```python
#print the result
print("T-statistics:",t_statistic)
print("p-value:",p_value)
print("critical value:",critical_value)
```

```python
if p_value<alpha:
    print("there is a significant difference in spending score between male and female groups.")
else:
    print("there is no significant difference in spending score between male and female groups")
    
```

```python
if abs(t_statistic)>critical_value:
    print("there is a significant difference in spending score between male and female groups")
else:
    print("there is no significant difference in spending score between male and female groups")
```
