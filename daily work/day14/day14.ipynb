<h1>Montecarlo Simulation<h/1>

```python
import numpy as np
```

```python
#part 1
def roll_dice():
    #simulation of rolling a dice twice minumum number we get is 1, max is 6
    #add the results of tyhe two simulations  so possible values will be:
    #(1,1)or (1,2)...(6,6) sum will be3 bewtween 2 and 12
    #run multiple times to verify 
    return np.sum(np.random.randint(1,7,2))
print(roll_dice())
```

```python
#part 2
#someone approaches us saying I will give you 5 dollars if you get 7 and take 1 dollar if you get a number other than 7
#how do we know what will happen ?
#our own "monte carlo simulation "like function
def monte_carlo_simulation(runs=1000):
    results =np.zeros(2) #an array results[0] and results[1] initialized to two zeroes
    for _ in range(runs):
        if roll_dice()==7:
            results[0]+=1  #win count
        else:
            results[1]+=1 #lose count
    return results
#test 2-3 times and calculate how much you will versus lose 
print(monte_carlo_simulation())
print(monte_carlo_simulation())
print(monte_carlo_simulation())
#results may be favorable to us but was that by luck?
```

```python
#part 3
#now do it 1000 times
#take some time
results=np.zeros(1000)

for i in range(1000):
    results[i]=monte_carlo_simulation()[0]
print(results)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
ax.hist(results,bins=15,color="lime",ec="black")
ax.set_facecolor((1.0, 0.47, 0.42))
plt.show()
```

```python
#our win/loss
print(results.mean())   #general mean
print(results.mean()*5)   #what we will get as win on an average
print(results.mean()*4.75)  #just a marginal change in win reward se the impact
print(1000-results.mean())   #what we will pay on an average
print(results.mean()/1000)   #probability of the "we will win " result
```

```python
#the last probability should be close to the theoretical probability of gettion a 7 when we throw two dice (why? 1+6,2+5,3+4,5+2,6+1=6 out of 36 states)
```

```python
#investment parameters 
investment=100000
num_years=20
stock_weight=0.6
bond_weight=0.4
```

```python
#return and standard deviation for stocks and bonds
stock_return=0.15
stock_sd=0.60
bond_return=0.07
bond_sd=0.05
```

```python
#number of sumulatins
iterations=5555
```

```python
#simulate random returns for stocks and bonds
#np.random.normal: this function generates random returns for stocks and bonds
#for each year of the investment horixon in each simulation 
#arguments:mean (average retuern ),standard deviation (vilatility),and simulation size(number of years and simulations)
#stock_returns and bond _returns will be 5555*20 arrays
stock_returns=np.random.normal(stock_return,stock_sd,(iterations,num_years))
bond_returns=np.random.normal(bond_return,bond_sd,(iterations,num_years))
```

```python
stock_returns
```

```python
total_returns_pct=((stock_weight*stock_returns)+(bond_weight*bond_returns))*100
```

```python
#suppress scientific notation for numpy arrays
np.set_printoptions(suppress=True)
total_returns_pct
```

```python
#create a 5555 *20 array that will store cumultive investment values
#example: consider portfolio_values[0][0]
#suppose total returns for the first 5 out of 20 years were 15% ,0% -2%,59% and 7%
#portfolio_values[0][0]=100000*(1+0.15)=100000*1.5=115000
#portfolio_values[0][1]=115000*(1+0.15)=115000*1.5=115000
#portfolio_values[0][02]=115000*(1+0.15)=115000*1.5=11700
#portfolio_values[0][3]=112000*(1+0.15)=112000*1.5=179193
#portfolio_values[0][4]=179193*(1+0.15)=179193*1.5=191736
```

```python
portfolip_values=investment*(1+(total_returns_pct/100)).cumprod(axis=1)
```

```python
print("*** Portfolio Values ***")
print(portfolip_values)
```

```python
#calculate final portfolio values statistics
final_values=portfolip_values[:,-1]
print(f"summary of final portfolio values after {num_years}years")
print(f"mean:{final_values.mean():.2f}")
print(f"standard deviation {final_values.std():.2f}")
```

```python
#calculate percentiles (eg 25th 50th 100th percentiles)
percentiles=np.percentile(final_values,[25,50,100])
print(f"\n perentiles of final portfolio values:")
print(f"25th percentiles:{percentiles[0]:.2f}")
print(f"50th percentiles:{percentiles[1]:.2f}")
print(f"100th perentiles:{percentiles[2]:2f}")
```

```python
#plot it 
plt.figure(figsize=(15,6))
plt.title("total returns % simulation")
plt.plot(total_returns_pct)
plt.show()
```

<h2>Design a simple game:
Choose a number between 1 and 100.
If the sum of the digits of your chosen number exceeds 12, you win 3 dollars.
If the sum falls below the threshold, you lose 1 dollar.
The game will be repeated 1000 times to simulate a large number of trials.
<h/2>


```python
def generate_number():
     return np.random.randint(1,101)
print(generate_number())
```

```python
#find sum of digits
def getSum(n): 
    sum = 0
    for digit in str(n):  
      sum += int(digit)       
    return sum
   
```

```python

#part 2
def monte_carlo_simulation(runs=1000):
    results =np.zeros(2)
    for _ in range(runs):
        if getSum(generate_number())>12:
            results[0]+=1  #win count
        else:
            results[1]+=1  #lose count
    return results

#test 2-3 times and calculate how much you will versus lose 
print(monte_carlo_simulation())
print(monte_carlo_simulation())
print(monte_carlo_simulation())
```

```python
#part 3
#now do it 1000 times
#take some time

results=np.zeros(1000)
for i in range(1000):
    results[i]=monte_carlo_simulation()[0]
print(results)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()
plt.hist(results,bins=15,color="lime",ec="black")
ax.set_facecolor((1.0, 0.47, 0.42))
plt.show()
```

```python
#our win/loss
print(results.mean())   #general mean
print(results.mean()*5)   #what we will get as win on an average
print(results.mean()*4.75)  #just a marginal change in win reward se the impact
print(1000-results.mean())   #what we will pay on an average
print(results.mean()/1000)   #probability of the "we will win " result
```

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("Social_Network_Ads.csv")
x=dataset.iloc[:,[2,3]].values
```

```python
from collections import Counter
```

```python
#get gender frequencies 
#example:{"male":249,"female":362}
gender_counts=Counter(dataset["Gender"])
```

```python
#define a function to encode gender values 
def encode_gender(gender):
    return gender_counts[gender]/len(dataset)
```

```python
#apply the encoding function to the gender column 
dataset["Gender"]=dataset["Gender"].apply(encode_gender)
print(dataset.head())
```

```python
#update th efeature set with the encoded gender colum
x=dataset.iloc[:,[1,2,3]].values
print(x)
```

```python
y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier=GaussianNB()
classifier.fit(xtrain,ytrain)
ypred=classifier.predict(xtest)
print(ypred) #our pedicted values
print(ytest)  #actual values
```

```python
#how did we perform?
from sklearn.metrics import confusion_matrix,accuracy_score
ac=accuracy_score(ytest,ypred)
cm =confusion_matrix(ytest,ypred)
```

```python
#count actual and predicted purchases
actual_no_purchase=(ytest==0).sum()
actual_purchase=(ytest==1).sum()
predicted_no_purchase=(ypred==0).sum()
predicted_purchase=(ypred==1).sum()
```

```python
#create bar chart data 
labels=["Actual No purchase","actual purchase","predicted no of purchase","predicted purchase"]
values=[actual_no_purchase,actual_purchase,predicted_no_purchase,predicted_purchase]
```

```python
#create the bar chart
plt.figure(figsize=(8,6))
plt.bar(labels,values,color=['tab:orange',"white","green","blue"],ec="b")
plt.xlabel("purchase status")
plt.ylabel("count")
plt.title("actual vs predicted purchases")
plt.xticks(rotation=45)  #rotate x axis labels for better readability
plt.tight_layout()
plt.show()
```

```python
#confusion matrix heatmap
cm=confusion_matrix(ytest,ypred)
#create a new figure for the heatmap
import seaborn as sns
plt.figure(figsize=(8,6))
sns.heatmap(cm,annot=True,cmap="viridis",fmt="d")
plt.xlabel("predicted purchase")
plt.ylabel("actual purchase")
plt.title("confusion matrix")
plt.show()
```
