```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
import random
```

```python
df1=pd.read_csv("marathon_results_2015.csv")
df2=pd.read_csv("marathon_results_2016.csv")
df3=pd.read_csv("marathon_results_2017.csv")
```

```python
df=pd.concat([df1,df2,df3])
df.head()
```

```python
sample=df.sample(frac=0.10, replace=True, random_state=1)
sample.head()
```

```python
def systematic_sampling(df,step):
    indexes=np.arange(0,len(df),step=step)
    systematic_sample=df.iloc[indexes]
    return systematic_sample


systematic_sampling=systematic_sampling(df,7)
systematic_sampling.head()
```

```python
df_male=df[df["M/F"]=="M"]
df_female=df[df["M/F"]=="F"]
```

```python
males_s=df_male.sample(frac=0.10, replace=True, random_state=1)
female_s=df_female.sample(frac=0.10,replace=True,random_state=1)
female_s
```

```python
final_sample=pd.concat([males_s,female_s])
final_sample
```
