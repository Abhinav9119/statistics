```python




import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Read the dataset
df = pd.read_csv('hs_heights_pair.csv')

# Extract the heights of boys and girls
boys_heights = df['boys']
girls_heights = df['girls']

# Calculate mean and standard deviation for boys
mean_boys = np.mean(boys_heights)
std_boys = np.std(boys_heights)

# Calculate mean and standard deviation for girls
mean_girls = np.mean(girls_heights)
std_girls = np.std(girls_heights)

print(f"Boys: Mean = {mean_boys}, Standard Deviation = {std_boys}")
print(f"Girls: Mean = {mean_girls}, Standard Deviation = {std_girls}")

# Convert Z-scores to original values
z_score_boy = -1.79
z_score_girl = 0.48
boy_height = mean_boys + z_score_boy * std_boys
girl_height = mean_girls + z_score_girl * std_girls

# Define colors for the SD lines
colors = ['green', 'orange', 'red']

# Plot the distributions
plt.figure(figsize=(14, 6))

# Plot for boys
plt.subplot(1, 2, 1)
sns.histplot(boys_heights, kde=True, stat="density", linewidth=0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_boys, std_boys)
plt.plot(x, p, 'k', linewidth=2)
title = "Boys' Heights\nMean = {:.2f}, SD = {:.2f}".format(mean_boys, std_boys)
plt.title(title)

# Plot mean and standard deviation lines for boys
for i in range(1, 4):
    plt.axvline(mean_boys + i * std_boys, color=colors[i-1], linestyle='--')
    plt.axvline(mean_boys - i * std_boys, color=colors[i-1], linestyle='--')

# Plot the Z-score data point for the boy
plt.plot(boy_height, 0, 'o', color='darkred', markersize=10)

# Highlight area under the curve for the boy
x_fill = np.linspace(xmin, boy_height, 100)
p_fill = norm.pdf(x_fill, mean_boys, std_boys)
plt.fill_between(x_fill, 0, p_fill, where=(x_fill <= boy_height), color='yellow', alpha=0.5)

# Plot for girls
plt.subplot(1, 2, 2)
sns.histplot(girls_heights, kde=True, stat="density", linewidth=0)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean_girls, std_girls)
plt.plot(x, p, 'k', linewidth=2)
title = "Girls' Heights\nMean = {:.2f}, SD = {:.2f}".format(mean_girls, std_girls)
plt.title(title)

# Plot mean and standard deviation lines for girls
for i in range(1, 4):
    plt.axvline(mean_girls + i * std_girls, color=colors[i-1], linestyle='--')
    plt.axvline(mean_girls - i * std_girls, color=colors[i-1], linestyle='--')

# Plot the Z-score data point for the girl
plt.plot(girl_height, 0, 'o', color='darkred', markersize=10)

# Highlight area under the curve for the girl
x_fill = np.linspace(xmin, girl_height, 100)
p_fill = norm.pdf(x_fill, mean_girls, std_girls)
plt.fill_between(x_fill, 0, p_fill, where=(x_fill <= girl_height), color='yellow', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
#q-q plot
stats.probplot(marks,fit=stats.norm,plot=plt)
plt.title("q-q plot of exam marks" )
plt.xlabel("theoretical quantitles")
plt.ylabel("samplt quantles")
plt.grid(True)
plt.show()
```

```python
#interpretation
if shapiro_test.pvalue<0.05:
    print("the data likely does not follow a normal distribution")
else:
    print("the data may be normally distributed but the qq plot can provide further insights.")
    
```

<h1>central limit theorem<h/1>

```python
from scipy.stats import shapiro
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df=pd.read_csv('diabetes.csv')
df.head()
```

```python
#central limit theorem
sample_means=[]
n_samples=1000
sample_size=30
```

```python
for _ in range(n_samples):
    sample=df['Age'].sample(n=sample_size,replace=True)
    sample_means.append(sample.mean())
```

```python
#plot distribution of sample means
plt.hist(sample_means,bins=30,edgecolor="k",alpha=0.7)
plt.title("distribution of sample means (n=30)")
plt.xlabel("sample mean")
plt.ylabel("frequency")
plt.show()
```

```python
#fenerate a q-q plot for the sample means
stats.probplot(sample_means,dist='norm',plot=plt)
plt.title("q-q plot of sample means")
plt.show()
```

```python
#perform the shapiro-wilk test on the sample means
shapiro_test=stats.shapiro(sample_means)
print("sahpiro - wilk p-value:",shapiro_test.pvalue)
```
