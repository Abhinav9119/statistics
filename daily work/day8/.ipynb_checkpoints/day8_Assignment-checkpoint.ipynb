```python
import pandas as pd
import numpy as np

# Generate data
exam_ids = np.arange(1, 100001)
marks = np.random.randint(0, 201, size=100000)

# Create the DataFrame
df = pd.DataFrame({"Exam Registration ID": exam_ids, "Marks": marks})

# Calculate the mean and standard deviation
mean_marks = df["Marks"].mean()
std_marks = df["Marks"].std()

# Calculate your Z-score
your_mark = df.loc[df["Exam Registration ID"] == 9999, "Marks"].values[0]
z_score = (your_mark - mean_marks) / std_marks

# Find your percentile using the Z-table
# Assuming a normal distribution, we'll find the area to the left of your Z-score
percentile = 100 * (1 - norm.cdf(z_score))

print(f"Your Z-score: {z_score:.4f}")
print(f"Your percentile: {percentile:.2f}%")
```
