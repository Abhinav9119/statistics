```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("weight-height.csv")
df
```

```python
male_df=df.loc[df.Gender=="Male"]
male_df
```

```python
female_df=df.loc[df.Gender=="Female"]
female_df
```

```python
def identify_outlier(data):
    q1=np.percentile(data,25)
    q3=np.percentile(data,75)
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    return np.where((data<lb)|(data>ub))[0]
```

```python
#box plot

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxplot(y=male_df.Height)
plt.title("male height")


plt.subplot(1,2,2)
sns.boxplot(y=female_df.Height)
plt.title('female height')
```

```python


#outlier detection
male_outlier=identify_outlier(male_df.Height)

print("outlier in male height using iqr method")
male_df.iloc[male_outlier]

```

```python
female_outlier=identify_outlier(female_df.Height)

print("outlier in female height using iqr method")
print(female_df.iloc[female_outlier])
```

```python

#box plot for weight

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
sns.boxplot(y=male_df['Weight'])
plt.title("Male Weight")

plt.subplot(1,2,2)
sns.boxplot(y=female_df['Weight'])
plt.title('Female Weight')
```

```python


# Outlier detection using weight
male_outler_weight=identify_outlier(male_df['Weight'])
print("Outlier in male weight using iqr method")
male_df.iloc[male_outler_weight]
```

```python

# Outlier detection using weight
female_outlier_weight= identify_outlier(female_df['Weight'])

print('Outlier in female weight using iqr method')
female_df.iloc[female_outlier_weight]
```
