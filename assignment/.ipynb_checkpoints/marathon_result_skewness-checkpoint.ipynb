```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read datasets
df1=pd.read_csv("marathon_results_2015.csv")
df2=pd.read_csv("marathon_results_2016.csv")
df3=pd.read_csv("marathon_results_2017.csv")
```

```python
#combining data
df=pd.concat([df1,df2,df3])
df
```

```python
df.info()
```

```python
df["Official Time"].describe()
```

```python
#df["Official Time"]=pd.to_time(df["Official Time"])

df['Official Time'] = pd.to_timedelta(df['Official Time']).dt.total_seconds()
```

```python
df.info()
```

```python
df
```

```python
df["Official Time"].describe()
```

```python


mean=df["Official Time"].mean()
median=df["Official Time"].median()
std=df["Official Time"].std()
var=df["Official Time"].var()
skew=df["Official Time"].skew()
kurtosis=df["Official Time"].kurtosis()
cv=std/mean
```

```python
print("additional statistics for the tip column:")
print(f"mean:{mean}\nMedian:{median}\nstandart deviation:{std}\nvariance:{var}\nskewness:{skew}\nkurtosis:{kurtosis}\ncoefficient of variation:{cv}%")
```

```python
#Ploting the distribution of the tip columns
plt.figure(figsize=(14,6))

#histogram
plt.subplot(1,2,1) #number of rows number of columns current subplot number
sns.histplot(df["Official Time"],bins=20,kde=True,color="magenta")
plt.title("distribution official time")
plt.xlabel("official time")
plt.ylabel("frequency")




#box plot 
plt.subplot(1,2,2)
sns.boxplot(y=df["Official Time"],color="magenta")
plt.title('box plot of official time')
plt.xlabel ("top amount")
#show the plot 
plt.tight_layout()
plt.show()
```
